```python
df_group_one = df[['drive-wheels', 'body-style', 'price']]
```

```python
df_group_one = df_group_one.groupby(['drive-wheels'], as_index = False).mean()
df_group_one
```

```python
df_group_one1 = df[['drive-wheels', 'body-style', 'price']]
df_group_one1 = df_group_one1.groupby(['drive-wheels', 'body-style'], as_index = False).mean()
df_group_one1
```

```python
grp_pivot = df_group_one1.pivot(index = 'drive-wheels', columns ='body-style')
grp_pivot
```

```python
grp_pivot = grp_pivot.fillna(0)
grp_pivot
```

```python
df_grp = df[['body-style', 'price']]
df_group_2 = df_grp.groupby(['body-style'], as_index = False).mean()
df_group_2
```

```python
plt.pcolor(grp_pivot, cmap = 'RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grp_pivot, cmap='RdBu')

#label names
row_labels = grp_pivot.columns.levels[1]
col_labels = grp_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grp_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grp_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
from scipy import stats
```
