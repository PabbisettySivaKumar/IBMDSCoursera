```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
#REPLACING '?' WITH 'NaN'
```

```python
import numpy as np
df.replace('?', np.nan, inplace = True)
df.head()
```

```python
#FOR MISSING DATA
```

```python
missing_data = df.isnull()
missing_data.head()
```

```python
#COUNT OF THE COLOUMNS
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print('')
```

```python
#REPLACING THE NaN WITH MEAN VALUES
```

```python
avg_norma_loss = df['normalized-losses'].astype('float').mean(axis = 0)
print('Average_normalized_losses:', avg_norma_loss)
```

```python
df['normalized-losses'].replace(np.nan, avg_norma_loss, inplace = True)
df.head()
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis = 0)
print('Average_stroke:', avg_stroke)
```

```python
df['stroke'].replace(np.nan, avg_stroke, inplace = True)
df.head()
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis = 0)
print('Average_horsepower:', avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace = True)
df.head()
```

```python
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis = 0)
print('Average_peak-rpm:', avg_peak_rpm)
```

```python
df['peak-rpm'].replace(np.nan , avg_peak_rpm, inplace = True)
df.head()
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

```python
#DROPING THE TARGET NaN VALUES
```

```python
df.dropna(subset = ['price'], axis = 0, inplace = True)
df.reset_index(drop =  True, inplace = True)
df.head()
```

```python
df.dtypes
```

```python
#CONVERTING THE COLUMN TYPES INTO CORRECT FORMAT TYPES
```

```python
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype('float')
df['normalized-losses'] = df['normalized-losses'].astype('int')
df['price'] = df['price'].astype('float')
df['peak-rpm'] = df['peak-rpm'].astype('float')
```

```python
df.dtypes
```

```python
#DATA STANDARDIZATION, COMMON FORMAT
```

```python
df['city-L/100km'] = 235/df['city-mpg']
df.head()
```

```python
df['highway-mpg'] = 235/df['highway-mpg']
df.rename(columns = {'highway-mpg' : 'highway-L/100 km'}, inplace = True)
df.head()
```

```python
#Normalization
```

```python
df['length'] = df['length']/df['length'].max()
df.head()
```

```python
df['width'] = df['width']/df['width'].max()
```

```python
df['height'] = df['height']/df['height'].max()
df.head()
```

```python
#DATA BINNING
```

```python
df['horsepower'] = df['horsepower'].astype(int)
df.dtypes
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepowerbins')
```

```python
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
group_names = ['low', 'medium', 'high']
```

```python
df['horse_power_binned'] = pd.cut(df['horsepower'], bins, labels = group_names, include_lowest = True)
df[['horsepower', 'horse_power_binned']].head()
```

```python
df['horse_power_binned'].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.bar(group_names, df['horse_power_binned'].value_counts())

plt.pyplot.xlabel('horse power')
plt.pyplot.ylabel('count')
plt.pyplot.title('horse power bins')
```

```python
#CRETING DUMMY VALUES & NEW COLUMNS
```

```python
df.columns
```

```python
dummies = pd.get_dummies(df['fuel-type'])
dummies.head()
```

```python
dummies.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummies.head()
```

```python
df = pd.concat([df, dummies], axis = 1)
df.head()
```

```python
#DROPPING ORIGINAL COLUMN
```

```python
df.drop('fuel-type', axis = 1, inplace = True)
```

```python
df.head()
```

```python
dummies1 = pd.get_dummies(df['aspiration'])
dummies1.head()
```

```python
dummies1.rename(columns = {'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace = True)
dummies1.head()
```

```python
df = pd.concat([df, dummies1], axis = 1)
df.head()
```

```python
df.drop('aspiration', axis = 1, inplace = True)
df.head()
```

```python
#SAVING TO NEW CSV
```

```python
df.to_csv('clean_DataSet.csv')
```
